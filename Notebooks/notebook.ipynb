Problema do negócio

Requisitos(KPI's)

Precisamos analisar os KPI dos dados sobre a venda de pizzas para obter insights sobre a performance de vendas.

1. Venda Total: A soma do total de todos pedidos de pizza.

2. Média de valor dos pedidos: A média do valor de pedido, calculado dividindo o total de vendas pelo total de pedidos.

3. Quantidade total de pizzas vendidas: A soma total da quantidade de pizzas vendidas no período.

4. Quantidade total de pedidos: A soma total da quantidade de pedidos no período.

5. Quantidade média de pizzas por pedido: A média do número de pizzas por pedido, calculado dividindo 
o total de pizzas vendidas pelo total de pedidos.


Requisitos(Gráficos)

Gostaríamos de vizualizar vários aspectos dos dados sobre venda de pizzas para obter insights e entender
suas principais tendências.

1. Tendência diária para total de pedidos: Criar um gráfico de barras que mostre a tendência diária do total de  pedidos
durante um período específico. Esse gráfico nos ajudará a identificar qualquer padrão ou flutuações no volume de pedidos diários.

2. Tendência(Mensal/Semanal/Hora) do total de pedidos: Criar um gráfico de linhas que mostre a tendencia do total de pedidos pelos períodos mencionados. Esse gráfico nos ajudará a identificar horários de pico e horários com número alto de pedidos.

3. Porcentagem de vendas por categoria de pizza: Criar um gráfico de pizza que mostre a distribuição de vendas por categorias de pizza. Esse gráfico nos fornecerá insights sobre a popularidade das várias categorias de pizza e suas 
contribuições para o total de vendas.

4. Porcentagem de vendas por tamanho de pizza: Criar um gráfico de pizza que represente a porcentagem de vendas dos tamanhos de pizzas. Esse gráfico vai nos ajudar a entender as preferencias dos clientes em relação ao tamanho da pizza e
o seu impacto nas vendas.

5. Total de pizzas vendidas por categoria: Criar um gráfico de funil que represente o total de pizzas vendidas por categoria. Esse gráfico vai nos ajudar a comparar a performance de vendas de cada categoria.

6. Top 5 das pizzas mais vendidas, por vendas, quantidade total e total de pedidos: Criar um gráfico de barras que destaque os 5 tipos de pizza mais vendidos, baseado nas vendas, quantidade total e total de pedidos. Esse gráfico nos ajudará a identificar as opções mais populares.

7. Os 5 tipos menos vendidos, por vendas, quantidade total e total de pedidos: Criar um gráfico de barras que destaque os 5 tipos de pizza menos vendidos, baseado nas vendas, quantidade total e total de pedidos. Esse gráfico nos ajudará a identificar as opções menos populares.
